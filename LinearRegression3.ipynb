```python
from sklearn.linear_model import LinearRegression # train
from sklearn.model_selection import train_test_split 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\M7MD\Desktop\Cellphone.csv")
```

```python
pd.set_option('display.max_rows', None)
```

```python
data = data.drop(['Product_id'] , axis = 1  )
```

```python
data
```

```python
x = data[['Sale','weight','resoloution','ppi','cpu core','cpu freq','internal mem','ram','RearCam','Front_Cam','battery','thickness']]
y = data['Price']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3,random_state = 101)
```

```python
model = LinearRegression() 
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
```

```python
plt.scatter(y_test,prediction)
```
